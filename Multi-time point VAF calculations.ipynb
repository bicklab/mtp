```python
def calculate_vaf(bam_file, chrom, pos, ref, alt):
    samfile = pysam.AlignmentFile(bam_file, "rb")
    ref_count = 0
    alt_count = 0

    for pileupcolumn in samfile.pileup(chrom, pos - 1, pos, stepper='samtools', truncate=True):
        for pileupread in pileupcolumn.pileups:
            if not pileupread.is_del and not pileupread.is_refskip:
                # query position is None if is_del or is_refskip is set.
                base = pileupread.alignment.query_sequence[pileupread.query_position]
                if base == ref:
                    ref_count += 1
                elif base == alt:
                    alt_count += 1

    samfile.close()

    total_count = ref_count + alt_count
    if total_count > 0:
        vaf = alt_count / total_count
    else:
        vaf = 0
    print(alt_count)
    print(ref_count)
    return vaf
```

```python
def create_tsv_content(samples, prefix):
    """
    Create TSV content for a given set of sample names and a file path prefix.

    Parameters:
    samples (list of str): List of sample names.
    prefix (str): The file path prefix for all files.

    Returns:
    str: The content of the TSV file.
    """
    tsv_lines = [f"{prefix}{sample}.bam\t{prefix}{sample}.bam.bai" for sample in samples]
    return "\n".join(tsv_lines)

# Example usage
file_prefix = "gs://bicklab-main-storage/Data/vantage/9700/bam/"
tsv_content = create_tsv_content(list(unique_samples), file_prefix)

with open('mtp_bam_files_list.tsv', 'w') as file:
    file.write(tsv_content)
```
